```python
# Function to process a DataFrame by creating two new columns 'window1' and 'window2'
def process(df):
    # Concatenate values from specific columns to create a string in the format 'col1:col2-col3'
    df['window1'] = df.iloc[:, 0].astype(str) + ':' + df.iloc[:, 1].astype(str) + '-' + df.iloc[:, 2].astype(str)
    df['window2'] = df.iloc[:, 3].astype(str) + ':' + df.iloc[:, 4].astype(str) + '-' + df.iloc[:, 5].astype(str)

    # Select specific columns ('window1', 'window2', 'NPMI.x', 'NPMI.y') and create a new DataFrame
    df_m = df[['window1', 'window2', 'NPMI.x', 'NPMI.y']]
    return df_m

# Apply the process function to a DataFrame wt_unique_contacts and sd_unique_contacts
wt_unique_contacts_m = process(wt_unique_contacts)
sd_unique_contacts_m = process(sd_unique_contacts)

# Function to process a DataFrame by creating a new column 'window'
def process2(df):
    # Concatenate values from specific columns to create a string in the format 'col1:col2-col3'
    df['window'] = df.iloc[:, 0].astype(str) + ':' + df.iloc[:, 1].astype(str) + '-' + df.iloc[:, 2].astype(str)

    # Select only the 'window' column and create a new DataFrame
    df_window = df[['window']]
    return df_window
```

```python
# Function to find intersecting contacts between specified genomic regions and contact data
def contacts_in_region(region):
    """
    Takes a table with Ensemble IDs and intersects it with the chosen data to get the Ensemble 
    IDs for the respective regions, removes the duplicated genes from the final table.
    """
    import pybedtools  # Library for working with genomic intervals
    import pandas as pd

    # Process the input DataFrame to create a 'window' column
    regions_window = process2(region) 
    regions_windows_array = regions_window['window'].values
    print(regions_windows_array)
    # Query the contact DataFrames to find rows where both 'window1' and 'window2' are in the specified regions
    regions_vs_wt_contacts_DEGs = wt_unique_contacts_m.query("(window1 in @regions_windows_array) & (window2 in @regions_windows_array)")
    regions_vs_sd_contacts_DEGs = sd_unique_contacts_m.query("(window1 in @regions_windows_array) & (window2 in @regions_windows_array)")

    # Optionally save the results to files
    # regions_windows_vs_wt_contacts_DEGs.to_csv('./randon_regions_Test/' + 'test' + '.txt', sep='\t', index=False, header=True)
    # regions_windows_vs_sd_contacts_DEGs.to_csv('./randon_regions_Test/' + 'test2' + '.txt', sep='\t', index=False, header=True)

    # Return the resulting DataFrames
    return regions_vs_wt_contacts_DEGs, regions_vs_sd_contacts_DEGs
```
